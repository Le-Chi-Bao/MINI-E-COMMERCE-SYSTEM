### T√°ch Resolution th√†nh Width v√† Height, ƒë·∫£m b·∫£o Width >= Height

```python
def process_resolution(df):
    """T√°ch Resolution th√†nh Width v√† Height, ƒë·∫£m b·∫£o Width >= Height"""
    df_processed = df.copy()
    
    # ƒê·∫¢M B·∫¢O RESET INDEX
    df_processed = df_processed.reset_index(drop=True)
    
    if 'Resolution' in df_processed.columns:
        # T√°ch Resolution
        resolution_split = df_processed['Resolution'].str.split('x', expand=True)
        df_processed[['Res_Width', 'Res_Height']] = resolution_split.astype('float64')
        
        # Chu·∫©n h√≥a: Width lu√¥n >= Height
        width_height = df_processed[['Res_Width', 'Res_Height']]
        df_processed['Res_Width'] = width_height.max(axis=1)
        df_processed['Res_Height'] = width_height.min(axis=1)
        
        # X√≥a c·ªôt g·ªëc
        df_processed = df_processed.drop('Resolution', axis=1)
    
    return df_processed

# S·ª≠ d·ª•ng v·ªõi reset index
X_train = process_resolution(X_train)
X_test = process_resolution(X_test)
```

```python
def handle_binary_features(df, binary_map=None):
    """
    X·ª≠ l√Ω binary features encoding
    """
    if binary_map is None:
        binary_map = {
            'has_telephoto': {'Kh√¥ng c√≥ camera tele': 0, 'C√≥ camera tele': 1},
            'has_ultrawide': {'Kh√¥ng c√≥ camera si√™u r·ªông': 0, 'C√≥ camera si√™u r·ªông': 1},
            'has_ois': {'Kh√¥ng c√≥ ch·ªëng rung OIS': 0, 'C√≥ ch·ªëng rung OIS': 1},
            'has_warranty': {'Kh√¥ng c√≥ b·∫£o h√†nh': 0, 'C√≥ b·∫£o h√†nh': 1}
        }
    
    df_processed = df.copy()
    
    print("\n=== X·ª¨ L√ù BINARY FEATURES ===")
    
    for feature, mapping in binary_map.items():
        if feature in df_processed.columns:
            missing_before = df_processed[feature].isnull().sum()
            df_processed[feature] = df_processed[feature].map(mapping).fillna(0).astype(int)
            print(f"{feature}: {missing_before} missing ‚Üí 0")
    
    return df_processed

X_train = handle_binary_features(X_train)
X_test = handle_binary_features(X_test)
```

## X·ª¨ L√ù OUTLIERS & MISSING NUMERIC THEO NG∆Ø·ª†NG TH·ª∞C T·∫æ

```python
def handle_noise_with_median(df, threshold_rules):
    """
    X·ª≠ l√Ω noise/outliers b·∫±ng median theo threshold
    """
    df_processed = df.copy()
    
    for feature, (low, high) in threshold_rules.items():
        if feature in df_processed.columns:
            outliers_mask = (df_processed[feature] < low) | (df_processed[feature] > high)
            outliers_count = outliers_mask.sum()
            
            if outliers_count > 0:
                median_clean = df_processed[~outliers_mask][feature].median()
                df_processed.loc[outliers_mask, feature] = median_clean
                print(f"   {feature}: {outliers_count} outliers ‚Üí median({median_clean:.2f})")
    
    return df_processed

def handle_missing_with_median(df):
    """
    X·ª≠ l√Ω missing values b·∫±ng median
    """
    df_processed = df.copy()
    numeric_features = df_processed.select_dtypes(include=['int64', 'float64']).columns.tolist()
    
    for feature in numeric_features:
        missing_count = df_processed[feature].isnull().sum()
        if missing_count > 0:
            median_clean = df_processed[feature].median()
            df_processed[feature] = df_processed[feature].fillna(median_clean)
            print(f"   {feature}: {missing_count} missing ‚Üí median({median_clean:.2f})")
    
    return df_processed

# H√ÄM X·ª¨ L√ù TARGET (y_train)
def handle_target_noise(y_series, threshold=70000000):
    """
    X·ª≠ l√Ω noise trong target variable b·∫±ng median
    """
    y_processed = y_series.copy()
    outliers_mask = y_processed > threshold
    outliers_count = outliers_mask.sum()
    
    if outliers_count > 0:
        median_clean = y_processed[~outliers_mask].median()
        y_processed[outliers_mask] = median_clean
        print(f"   DiscountedPrice: {outliers_count} outliers ‚Üí median({median_clean:,.0f} VND)")
    
    return y_processed

print("1. X·ª¨ L√ù NOISE/OUTLIERS:")

# X·ª≠ l√Ω noise cho features
noise_rules = {
    'ScreenSize': (4, 8),
    'NumberOfReview': (0, 1000),
    'main_camera_mp': (5, 200),
    'num_cameras': (1, 5),
    'Res_Width': (720, 3840),
    'Res_Height': (1280, 2160)
}

X_train = handle_noise_with_median(X_train, noise_rules)
X_test = handle_noise_with_median(X_test, noise_rules)

# N√™n: Log transform tr∆∞·ªõc ‚Üí x·ª≠ l√Ω outliers tr√™n log scale.
y_train = np.log1p(y_train)
y_test = np.log1p(y_test)

# X·ª≠ l√Ω noise cho target
y_train = handle_target_noise(y_train)
y_test = handle_target_noise(y_test)

print("\n2. X·ª¨ L√ù MISSING VALUES:")

# X·ª≠ l√Ω missing values cho features
X_train = handle_missing_with_median(X_train)
X_test = handle_missing_with_median(X_test)

#num_cameras kh√¥ng c√≥ outliers nh∆∞ng c√≥ missing values.
```

## Feature Engrineering

### Th√™m c√°c Feature Views (5 Feature Views)

```python
def create_all_features(df, discounted_prices=None):
    """
    Create all features needed for 5 Feature Views
    df: features (kh√¥ng c√≥ DiscountedPrice)
    discounted_prices: series ch·ª©a gi√° (n·∫øu ƒë√£ t√°ch ra)
    """
    df_processed = df.copy()
    
    # ==================== X·ª¨ L√ù GI√Å ====================
    if discounted_prices is not None:
        # N·∫øu truy·ªÅn v√†o prices ri√™ng
        df_processed['DiscountedPrice'] = discounted_prices
        print("‚úÖ Using provided discounted prices")
    elif 'DiscountedPrice' in df_processed.columns:
        # N·∫øu gi√° v·∫´n c√≤n trong df
        print("‚úÖ Using DiscountedPrice from dataframe")
    else:
        # N·∫øu kh√¥ng c√≥ gi√°, d√πng m·∫∑c ƒë·ªãnh
        print("‚ö†Ô∏è  No price data found, using default value 8000000")
        df_processed['DiscountedPrice'] = 8000000
    
    # ==================== [PH·∫¶N C√íN L·∫†I GI·ªÆ NGUY√äN] ====================
    # DISPLAY VIEW FEATURES
    df_processed['PPI'] = np.sqrt(
        df_processed['Res_Width']**2 + df_processed['Res_Height']**2
    ) / df_processed['ScreenSize']
    
    df_processed['total_resolution'] = df_processed['Res_Width'] * df_processed['Res_Height']
    
    # CAMERA VIEW FEATURES
    df_processed['camera_feature_count'] = (
        df_processed['has_telephoto'] + 
        df_processed['has_ultrawide'] + 
        df_processed['has_ois']
    )
    
    df_processed['camera_score'] = (
        df_processed['main_camera_mp'] * 0.4 +
        df_processed['num_cameras'] * 0.3 + 
        df_processed['camera_feature_count'] * 0.3
    )
    
    # RATINGS VIEW FEATURES
    main_camera_clean = df_processed['main_camera_mp'].fillna(0)
    num_cameras_clean = df_processed['num_cameras'].fillna(1)
    feature_count_clean = df_processed['camera_feature_count'].fillna(0)
    
    max_camera_mp = max(main_camera_clean.max(), 50.0)
    max_num_cameras = max(num_cameras_clean.max(), 5.0)
    max_features = max(feature_count_clean.max(), 3.0)
    
    camera_quality = (
        (main_camera_clean / max_camera_mp).clip(0, 1) * 0.4 +
        (num_cameras_clean / max_num_cameras).clip(0, 1) * 0.3 +
        (feature_count_clean / max_features).clip(0, 1) * 0.3
    )
    df_processed['camera_rating'] = (camera_quality * 4 + 1).round(1)
    
    # Display performance score
    ppi_clean = df_processed['PPI'].fillna(300)
    resolution_clean = df_processed['total_resolution'].fillna(2000000)
    screen_clean = df_processed['ScreenSize'].fillna(6.0)
    
    ppi_90 = float(ppi_clean.quantile(0.9)) if len(ppi_clean) > 0 else 600
    resolution_90 = float(resolution_clean.quantile(0.9)) if len(resolution_clean) > 0 else 8000000
    screen_90 = float(screen_clean.quantile(0.9)) if len(screen_clean) > 0 else 7.5
    
    display_score = (
        (ppi_clean / ppi_90).clip(0, 1) * 40 +
        (resolution_clean / resolution_90).clip(0, 1) * 40 +
        (screen_clean / screen_90).clip(0, 1) * 20
    )
    df_processed['display_score'] = display_score.round(1)
    
    # Popularity score
    if 'NumberOfReview' in df_processed.columns:
        reviews_clean = df_processed['NumberOfReview'].fillna(0).clip(lower=0)
        current_max = float(reviews_clean.max())
        if current_max > 0:
            df_processed['popularity_score'] = (
                (np.log1p(reviews_clean) / np.log1p(current_max)) * 100
            ).round(1)
        else:
            df_processed['popularity_score'] = 0
    
    # Overall score
    score_components = []
    if 'camera_rating' in df_processed.columns:
        camera_component = (df_processed['camera_rating'] - 1) / 4 * 100 * 0.3
        score_components.append(camera_component)
    
    if 'display_score' in df_processed.columns:
        display_component = df_processed['display_score'] * 0.4
        score_components.append(display_component)
    
    if 'popularity_score' in df_processed.columns:
        popularity_component = df_processed['popularity_score'] * 0.3
        score_components.append(popularity_component)
    
    if score_components:
        total_score = sum(score_components)
        df_processed['overall_score'] = total_score.round(1)
    else:
        df_processed['overall_score'] = 0
    
    # ==================== VALUE VIEW FEATURES (TARGET COLUMNS) ====================
    price_clean = df_processed['DiscountedPrice'].fillna(8000000)
    
    # Premium detector
    df_processed['is_premium'] = (price_clean > 15000000).astype(int)
    
    # Price segment (0: budget, 1: mid_range, 2: premium)
    conditions = [
        price_clean <= 8000000,
        price_clean <= 15000000, 
        price_clean > 15000000
    ]
    choices = [0, 1, 2]
    df_processed['price_segment'] = np.select(conditions, choices, default=1)
    
    # Value for money score
    camera_max = max(float(df_processed['camera_score'].max()), 1.0)
    feature_value = (
        (df_processed['camera_score'].fillna(0) / camera_max * 50) + 
        (df_processed['display_score'].fillna(0) / 100 * 50)
    ).clip(0, 100)
    
    price_in_millions = price_clean / 1000000
    valid_price_mask = price_in_millions > 0.1
    
    value_scores = np.zeros(len(df_processed))
    if valid_price_mask.any():
        value_scores[valid_price_mask] = (
            feature_value[valid_price_mask] / price_in_millions[valid_price_mask]
        ).round(2)
    
    if value_scores.max() > 0:
        value_scores = (value_scores / value_scores.max() * 10).round(2)
    
    df_processed['value_score'] = value_scores
    
    # Fill any remaining NaN values with 0
    numeric_columns = df_processed.select_dtypes(include=[np.number]).columns
    df_processed[numeric_columns] = df_processed[numeric_columns].fillna(0)
    
    return df_processed
```

### Train Model

#### 1. üöÄ SETUP & PREPARATION

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.metrics import mean_squared_error, accuracy_score, classification_report
from sklearn.model_selection import cross_val_score
import matplotlib.pyplot as plt
import seaborn as sns

# Kh·ªüi t·∫°o models
models = {
    'smart_recommender': RandomForestRegressor(n_estimators=100, random_state=42),
    'value_detector': RandomForestClassifier(n_estimators=100, random_state=42),
    'camera_predictor': RandomForestRegressor(n_estimators=100, random_state=42)
}
```

#### 2. üìä DEFINE FEATURES & TARGETS

```python
# Feature sets cho t·ª´ng model
feature_sets = {
    'smart_recommender': [
        'ScreenSize', 'PPI', 'total_resolution',
        'camera_score', 'has_telephoto', 'has_ultrawide', 
        'popularity_score', 'value_score', 'price_segment',
        'has_warranty', 'NumberOfReview'
    ],
    'value_detector': [
        'value_score', 'price_segment', 'overall_score',
        'display_score', 'camera_rating', 'PPI',
        'ScreenSize', 'camera_score', 'main_camera_mp',
        'NumberOfReview'
    ],
    'camera_predictor': [
        'main_camera_mp', 'num_cameras', 'has_telephoto',
        'has_ultrawide', 'has_ois', 'camera_feature_count',
        'PPI', 'total_resolution', 'ScreenSize',
        'value_score', 'is_premium', 'NumberOfReview'
    ]
}

# Target variables
targets = {
    'smart_recommender': 'overall_score',
    'value_detector': 'is_premium', 
    'camera_predictor': 'camera_rating'
}
```

#### 3. üîß TRAINING FUNCTION

```python
def train_model(model_name, X_train, X_test, y_train, y_test):
    """Train v√† evaluate model"""
    print(f"üéØ Training {model_name.upper()}...")
    print("-" * 50)
    
    # L·∫•y features v√† target
    features = feature_sets[model_name]
    target = targets[model_name]
    
    # Ki·ªÉm tra features t·ªìn t·∫°i
    missing_features = [f for f in features if f not in X_train.columns]
    if missing_features:
        print(f"‚ùå Missing features: {missing_features}")
        return None
    
    # Prepare data
    X_train_sel = X_train[features]
    X_test_sel = X_test[features]
    
    print(f"üìä Features: {len(features)}")
    print(f"üéØ Target: {target}")
    print(f"üìà Train size: {X_train_sel.shape}")
    print(f"üìä Test size: {X_test_sel.shape}")
    
    # Train model
    model = models[model_name]
    model.fit(X_train_sel, y_train)
    
    # Predictions
    y_pred = model.predict(X_test_sel)
    
    # Evaluation
    if model_name == 'value_detector':  # Classification
        accuracy = accuracy_score(y_test, y_pred)
        print(f"‚úÖ Accuracy: {accuracy:.4f}")
        print("\nüìã Classification Report:")
        print(classification_report(y_test, y_pred))
    else:  # Regression
        mse = mean_squared_error(y_test, y_pred)
        rmse = np.sqrt(mse)
        print(f"‚úÖ RMSE: {rmse:.4f}")
        print(f"‚úÖ MSE: {mse:.4f}")
    
    # Feature importance
    feature_importance = pd.DataFrame({
        'feature': features,
        'importance': model.feature_importances_
    }).sort_values('importance', ascending=False)
    
    print(f"\nüîù Top 5 Important Features:")
    print(feature_importance.head())
    
    print("=" * 50)
    return model, feature_importance
```

#### 4. üèÉ TRAIN ALL MODELS

```python
# Dictionary ƒë·ªÉ l∆∞u k·∫øt qu·∫£
trained_models = {}
feature_importances = {}

# Train t·ª´ng model
for model_name in models.keys():
    print(f"\n{'='*60}")
    print(f"üöÄ TRAINING {model_name.upper()}")
    print(f"{'='*60}")
    
    # L·∫•y target
    target_col = targets[model_name]
    
    # Ki·ªÉm tra target c√≥ t·ªìn t·∫°i kh√¥ng
    if target_col not in X_train_all.columns:
        print(f"‚ùå Target '{target_col}' not found in data")
        continue
        
    # Prepare target data
    y_train_model = X_train_all[target_col]
    y_test_model = X_test_all[target_col]
    
    # Train model
    result = train_model(model_name, X_train_all, X_test_all, y_train_model, y_test_model)
    
    if result:
        trained_models[model_name], feature_importances[model_name] = result
```

#### 5. üîç MODEL COMPARISON

```python
# So s√°nh hi·ªáu su·∫•t models
print("üìä MODEL PERFORMANCE SUMMARY")
print("=" * 60)

performance_data = []

for model_name in trained_models.keys():
    features = feature_sets[model_name]
    target = targets[model_name]
    
    X_train_sel = X_train_all[features]
    X_test_sel = X_test_all[features]
    y_train_model = X_train_all[target]
    y_test_model = X_test_all[target]
    
    model = trained_models[model_name]
    y_pred = model.predict(X_test_sel)
    
    if model_name == 'value_detector':
        score = accuracy_score(y_test_model, y_pred)
        metric = 'Accuracy'
    else:
        score = np.sqrt(mean_squared_error(y_test_model, y_pred))
        metric = 'RMSE'
    
    performance_data.append({
        'Model': model_name.upper(),
        'Target': target,
        'Metric': metric,
        'Score': f"{score:.4f}",
        'Features': len(features)
    })

performance_df = pd.DataFrame(performance_data)
print(performance_df)
```
